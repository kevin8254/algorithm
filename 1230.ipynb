```
#DFS
def enqueue(a, o):
    a.insert(0, o)

def dequeue(a):
    return a.pop()

g = { #graph:被搜尋的網路
    '1': {'n':['2','3','6'], 'v':0},  #n:neighbor(鄰居),v:visited(是否被訪問過)
    '2': {'n':['3'],'v':0},
    '3': {'n':['6'],'v':0},
    '4': {'n':['5'],'v':0},
    '5': {'n':[ ],'v':0},
    '6': {'n':['4'],'v':0}
}

def init(g):  #初始化、設定visited為0
    for i in g:
        g[i]['v'] = 0

def dfs(g,node): # 深度優先搜尋
    if g[node]['v']!=0:           #如果已訪問過，就不再訪問
        return
    print(node, '=> ',end = '')    #否則、印出節點
    g[node]['v'] = 1          #並設定為已訪問
    neighbors = g[node]['n']      #取出鄰居節點
    for n in neighbors:          #對於每個鄰居
        dfs(g,n)            #逐一進行訪問


print('dfs:', end = '')
init(g)
dfs(g,'1')   #呼叫深度優先搜尋。
print('')


```

```
#BFS
graph = {
    "1":["2","3","6"],
    "2":["3"], 
    "3":["6"],
    "4":["5"],
    "5":[],
    "6":["4"],      
}
def BFS(graph,s):
    queue = []
    queue.append(s)
    seen = set()
    seen.add(s)
    while(len(queue)>0):
        vertex = queue.pop(0)
        nodes = graph[vertex]
        for w in nodes:
            if w not in seen:
                queue.append(w)
                seen.add(w)
        print(vertex)
BFS(graph,'1')
```
