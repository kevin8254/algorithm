Selection sort (選擇排序)

```
def Selction_Sort(datas):                         #計算過程
    for i in range(len(datas)-1):                   #n 
        min = i                          #1
        for j in range(i+1, len(datas)):            #n+1
            if datas[j] < datas[min]:            #n
                min = j                  #n*(n+1)

        datas[i], datas[min] = datas[min], datas[i]      #1

        print(datas)                        #1
        print('---'*20)                      #0
    return datas                           # 1 


if __name__ == '__main__':
    import random                         #n
    a = list(range(16))                     #n
    random.shuffle(a)                        #n
    print(a)                            #1
    a = Selction_Sort(a)                    #1
    print(a)                          #O(n*n)
```
